```python
StandStill = []
RunWalk = []
Crawl = []
Climbupdown = []
Sit = []
LieDown=[]
Carried=[]
Stroller = []

def seperateclips(indnum,newlist):
    for index, row in dfal325sec.iterrows():
        if (dfal325sec.iloc[index,1] ==indnum):
            newlist.append(dfal325sec.iloc[index,2])
    print(indnum)
    print(newlist)
seperateclips(201.0,StandStill)
seperateclips(101.0,RunWalk)
seperateclips(102.0,RunWalk)
seperateclips(103.0,Crawl)
seperateclips(104.0,Climbupdown)
seperateclips(105.0,Climbupdown)
seperateclips(301.0,Sit)
seperateclips(302.0,LieDown)
seperateclips(303.0,Carried)
seperateclips(304.0,Stroller)




```

```python

XStand=[]
YStand=[]
ZStand=[]

XRunWalk=[]
YRunWalk=[]
ZRunWalk=[]

XCrawl=[]
YCrawl=[]
ZCrawl=[]

XClimbupdown=[]
YClimbupdown=[]
ZClimbupdown=[]

XSit=[]
YSit=[]
ZSit=[]

XLieDown=[]
YLieDown=[]
ZLieDown=[]

XCarried=[]
YCarried=[]
ZCarried=[]

XStroller=[]
YStroller=[]
ZStroller=[]

def XYZseperate(Compiled,X,Y,Z):
    Clips = Compiled[1]
    for i in Clips:
        X.append(i[0])
        Y.append(i[1])
        Z.append(i[2])
    
XYZseperate(StandStill,XStand,YStand,ZStand)
XYZseperate(RunWalk,XRunWalk,YRunWalk,ZRunWalk)
XYZseperate(Crawl,XCrawl,YCrawl,ZCrawl)
XYZseperate(Climbupdown,XClimbupdown,YClimbupdown,ZClimbupdown)
XYZseperate(Sit,XSit,YSit,ZSit)
XYZseperate(LieDown,XLieDown,YLieDown,ZLieDown)
XYZseperate(Carried,XCarried,YCarried,ZCarried)
XYZseperate(Stroller,XStroller,YStroller,ZStroller)
```

```python
import matplotlib.pyplot as plt
import numpy as np
TimePlot = list(range(600))
Timeskip=[]
for t in TimePlot:
    if t % 2 == 0:
        Timeskip.append(t)
        
def plotgraph(X,Y,Z,title):
    xNparry = np.array(X);
    yNparry = np.array(Y);
    zNparry = np.array(Z);
    
    plt.plot(Timeskip,xNparry,label='X')
    plt.plot(Timeskip,yNparry,label='Y')
    plt.plot(Timeskip,zNparry,label='Z')
    plt.axis([0,120,-1.5,2.0])
    plt.legend()
    plt.xlabel('Time(ms)')
    plt.ylabel('Acceleration(g)')
    
    plt.title(title)
    plt.show()
    
    
    
plotgraph(XStand,YStand,ZStand,"Stand")
plotgraph(XRunWalk,YRunWalk,ZRunWalk,"Run/Walk")
plotgraph(XCrawl,YCrawl,ZCrawl,"Crawl")
plotgraph(XClimbupdown,YClimbupdown,ZClimbupdown,"Climb Up & Down")
plotgraph(XSit,YSit,ZSit,"Sit")
plotgraph(XLieDown,YLieDown,ZLieDown,"Lie Down")
plotgraph(XCarried,YCarried,ZCarried,"Carried")
plotgraph(XStroller,YStroller,ZStroller,"Stroller")
```
